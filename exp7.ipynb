```python
import matplotlib.pyplot as plt
students = ["arav","diya","kabir","riya","arjun"]
score = [98,67,58,86,77]
plt.barh(students,score, color = "red")
plt.title("Students Score")
plt.xlabel("Score")
plt.ylabel("Students")
plt.show()
```

```python
import matplotlib.pyplot as plt
students=["Cristiano","Mbappe","Ramos"]
fees=[3000,3400,2400]
plt.pie(fees, labels=students, autopct='%1.1f%%')
plt.show()
```

```python
import plotly.express as px

students = ["Cristiano", "Mbappe", "Ramos"]
fees = [3000, 3400, 2400]

fig = px.bar(x=students, y=fees, title="Student Fees")
fig.write_html("student_fees_bar_chart.html")
fig.show()
```
