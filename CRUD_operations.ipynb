CREATE

```python
student = {}

student["Name"] = "Alice"
student["Age"] = 20
student["Grade"]= "A"
print(student)
```

READ

```python
print(student["Name"]) #read the key
print(student.get("age"))
print(student.get("grade"))
print(student.get("grade","Not available"))
```

UPDATE

```python
student["Name"] = "BOb" #update the key name, its value is bob
student["Age"] = "12" #update the key Age, its value is 12
student["Grade"] = "A" #update the key Grade, its value is A
print(student)
```

DELETE

```python
del student["Grade"] # else pop function can also be used
print(student)
```

REMOVED

```python
removed = student.pop("Age")
print(removed)
print(student)
```

NESTED DICTIONARY

```python
#create a nested dictionary with student data
student = {
    "Alice": {"Math": 85, "Science": 92,"English": 95},
    "Bob": {"Math": 78, "Science": 88,"English": 96},
    "Charlie": {"Math": 95, "Science": 90,"English":98}

}
student
```

For manipulation using dictionary convert the dictionary to dataframe

```python
import pandas as pd
data = pd.DataFrame(student) #pass the dictionary to data frame and specify an index
data
```

```python
data["Average"] = data.mean(axis=1)
data
```

```python
data["Math"] = data.mean(axis=0)
#axis = 0 denotes columnwise access to the dataframe
data
#The result is not a number
```

Data Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color = "skyblue",edgecolor = "black")
plt.title("Average Marks of the students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

How to visualise subjectwise trend

```python
import pandas as pd
import matplotlib.pyplot as plt

# Re-create the dataframe from the original student dictionary
student = {
    "Alice": {"Math": 85, "Science": 92,"English": 95},
    "Bob": {"Math": 78, "Science": 88,"English": 96},
    "Charlie": {"Math": 95, "Science": 90,"English":98}

}
data = pd.DataFrame(student)

data_transposed = data.transpose() # Transpose the dataframe

plt.figure(figsize=(8,5))
for subject in ["Math","Science","English"]:
  plt.plot(data_transposed.index, data_transposed[subject], marker ='o', label=subject) # Plot against the transposed index (students)
plt.title("Subject wise performance ")
plt.xlabel("Students") # Changed xlabel to "Students"
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Re-create the dataframe from the original student dictionary
student = {
    "Alice": {"Math": 85, "Science": 92,"English": 95},
    "Bob": {"Math": 78, "Science": 88,"English": 96},
    "Charlie": {"Math": 95, "Science": 90,"English":98}

}
data = pd.DataFrame(student)

data_transposed = data.transpose() # Transpose the dataframe

plt.figure(figsize=(8,5))
data_transposed[["Math","Science","English"]].plot(kind="box") # Plot box plots for each subject
plt.title("Marks distribution by Subject") # Changed title
plt.xlabel("Subjects") # Changed xlabel
plt.ylabel("Marks") # Changed ylabel
plt.show()
```
